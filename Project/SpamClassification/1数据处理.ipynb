```python
import os
import pandas as pd
```

# 索引文件读取

```python
def getLabel():
    dictMap = {'ham':0, 'spam':1}
    indexDict = {}
    with open('./trec06c/full/index','r') as f:
        for line in f:
            value,key = line.split(' ')
            key = key.replace('../data','').replace('\n','')
            indexDict[key] = dictMap[value.lower()]
            # break
    return indexDict

# getLabel()
```

# 单个文件数据读取

```python
def getEmailContent(filepath):
    contentDict = {'from':'','to':'','date':'','content':''}# 初始化内容为空
    
    with open(filepath,'r', encoding='gbk',errors='ignore') as f:
        isNone = False # 检测是否有空行，空行下面是正文
        
        for line in f:
            line = line.strip().replace('#','')
            
            if not line:
                isNone = True
            elif line.startswith('From:'):
                contentDict['from'] = line[5:].strip()
            elif line.startswith('To:'):
                contentDict['to'] = line[3:].strip()
            elif line.startswith('Date:'):
                contentDict['date'] = line[5:].strip()
            
            if isNone:
                contentDict['content'] += line
        
    return contentDict

# getEmailContent('./trec06c/data/000/000')
```

```python
# 遍历子目录,每个子目录生成一个文件
def getFolderAllFileContent(root):
    pathList = os.listdir(root) # 根目录下所有文件和文件夹
    indexDict = getLabel() # 得到文件的标签
    
    
    
    for path in pathList: # 遍历根目录
        folderPath = root+path+'/' # 子目录
        fileList = os.listdir(folderPath) # 子目录下所有文件和文件夹
        
        if not os.path.exists('./trec06c/pre'):
            os.mkdir('./trec06c/pre')
        
        fileName = './trec06c/pre/'+path # 子目录生成文件的文件名
        print('正在处理第'+fileName)
        
        for file in fileList: # 遍历子目录
            key = '/'+path+'/'+file
            fileContent = getEmailContent(folderPath+file) # 得到单个文件内容
            fileContent['label'] = str(indexDict[key]) # 在内容中添加标签
            
            with open(fileName,'a',encoding='utf8') as f:
                content = '##'.join(fileContent.values())+'\n' # 将字典拼接成字符串，写入文件
                f.write(content)
            # break
        
        # break
    print('处理完成')

getFolderAllFileContent('./trec06c/data/')
```

```python
def mergeData(folderPath):
    fileList = os.listdir(folderPath)
    
    if not os.path.exists('./datas'):
        os.mkdir('./datas')
    
    with open('./datas/data_clean.csv','w',encoding='utf8') as  wf:
        for file in fileList:
            filepath = './trec06c/pre/'+file
            print('正在合并',filepath)
            
            with open(filepath,'r',encoding='utf8',errors='ignore') as rf:
                for line in rf:
                    wf.write(line)
                    
    print('合并完成')
    
mergeData('./trec06c/pre/')
```
